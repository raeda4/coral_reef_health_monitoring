```python
import ee
import geemap

```

```python
ee.Authenticate()
ee.Initialize(project='ee-raedambach')
longitude = -149.56194
latitude = -17.00872
```

```python
reef = ee.Geometry.Point([longitude, latitude])
dataset = (ee.ImageCollection('COPERNICUS/S2')
  .filterBounds(reef)
  .filterDate('2024-01-01', '2024-01-02')
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))
  .median())
```

```python
# Compute NDWI (Water Quality Indicator)
ndwi = dataset.normalizedDifference(['B3', 'B8'])
```

```python
# Compute NDCRI (Coral Health Indicator)
ndcri = dataset.normalizedDifference(['B2', 'B4'])
```

```python
# Create a map
Map = geemap.Map()
Map.centerObject(reef, 10)
```

```python
# Visualization Parameters
ndwi_vis = {'min': -1, 'max': 1, 'palette': ['blue', 'white', 'red']}
ndcri_vis = {'min': -1, 'max': 1, 'palette': ['green', 'yellow', 'red']}
```

```python
# Add layers to the map
Map.addLayer(ndwi, ndwi_vis, 'NDWI - Water Quality')
Map.addLayer(ndcri, ndcri_vis, 'NDCRI - Coral Health')
```

```python
reef = ee.Geometry.Point([longitude, latitude])
dataset = (ee.ImageCollection('COPERNICUS/S2')
  .filterBounds(reef)
  .filterDate('2024-01-01', '2024-01-02')
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))
  .median())

Map.centerObject(reef, 10)
Map.addLayer(dataset, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000}, 'Sentinel-2')
Map
```
