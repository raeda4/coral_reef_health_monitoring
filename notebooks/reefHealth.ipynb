```python
import ee
import geemap

```

```python
ee.Authenticate()
ee.Initialize(project='ee-raedambach')
latitude = -21.303269
longitude = 151.547206

# Big Broadhurst Reef: -18.928790, 147.757198
# Teti'aroa, an atoll in French Polynesia: -149.56194, -17.00872
# Coral Sea Habitat Protection Zone: -19.38139542259404, 155.85508494416055
# Coral Sea: -21.303269, 151.547206
```

```python
reef = ee.Geometry.Point([longitude, latitude]).buffer(5000)
dataset = (ee.ImageCollection('COPERNICUS/S2_HARMONIZED')
  .filterBounds(reef)
#  .filterDate('2024-01-01', '2024-02-01')
  .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 50))
  )

print('Number of images in collection:', dataset.size().getInfo())
```

```python
#creating as a dataset
dataset = dataset.median()
```

```python

# Compute NDWI (Water Quality Indicator)
ndwi = dataset.normalizedDifference(['B3', 'B8'])
```

```python
# Compute NDCRI (Coral Health Indicator)
ndcri = dataset.normalizedDifference(['B2', 'B4'])
```

```python
# Create a map
Map = geemap.Map()
Map.centerObject(reef, 10)
```

```python
# Visualization Parameters
ndwi_vis = {'min': -1, 'max': 1, 'palette': ['blue', 'white', 'red']}
ndcri_vis = {'min': -1, 'max': 1, 'palette': ['green', 'yellow', 'red']}
```

```python
# Add layers to the map
Map.addLayer(ndwi, ndwi_vis, 'NDWI - Water Quality')
Map.addLayer(ndcri, ndcri_vis, 'NDCRI - Coral Health')
```
